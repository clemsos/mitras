### Load CSV

```python
meme_name="The_Voice"

# name files
meme_csv=meme_name+".csv"
gephi_nodes_path=meme_name+"_nodes.csv"
gephi_edges_path=meme_name+"_edges.csv"
words_file=meme_name+"_words.csv"
```

### Extract conversationel user graph

```python
with open(meme_csv, 'rb') as csvfile:
    memecsv=csv.reader(csvfile)
    memecsv.next() # skip headers
    
    edges=[]
    nodes=[]
    
    for row in memecsv:

         # extract text
        t=row[1]    
        
        # regexp extract tweet entities
        mentions,urls,hashtags,clean=minetweet.extract_tweet_entities(t)
        
        # add timestamp to generate a dynamic graph
        d=datetime.datetime.strptime(row[9], "%Y-%m-%dT%H:%M:%S")
        timestamp=datetime.datetime(d.year,d.month,d.day,d.hour,d.minute,d.second)
        
        # if row[0] not in nodes : nodes.append(row[0])
        
        for mention in mentions:
            edges.append((row[0],mention,timestamp))
            # if mention not in nodes : nodes.append(mention)
    
        # retweeted_uid
        if row[7] != "" : 
            edges.append((row[7],row[0],timestamp))        
            # if row[7] not in nodes : nodes.append(row[7])
        
    print "User diffusion : %d nodes, %d edges"%(len(nodes), len(edges))
```

### Plot Graph

```python
%pylab inline
import networkx as nx

G=nx.Graph()
G.add_edges_from([(e[0],e[1]) for e in edges])
G.name="diffusion graph"
# nx.draw(G)
```

### Extract time series

```python
with open(meme_csv, 'rb') as csvfile:
    memecsv=csv.reader(csvfile)
    memecsv.next() # skip headers
    
    dates=[]
    values={}
    
    for row in memecsv:
        
        d=datetime.datetime.strptime(row[9], "%Y-%m-%dT%H:%M:%S")
        day = datetime.datetime(d.year,d.month,d.day,d.hour,0,0) # round by day
        i=day.strftime("%s")
        if day not in dates : # collect values 
            values[i]=0    
        values[i]+=1
        dates.append(day)
```

### Plot time series

```python
%pylab inline

import matplotlib.pyplot as plt

# prepare data
graph_title=meme_name+" Volume of Tweets - Time Series"

# sort values by time
vy=[values[v] for v in sorted(values.keys())]
vx=[datetime.datetime.fromtimestamp(float(v)) for v in sorted(values.keys())]

# Create a figure
w=20 # width of the canvas
h=10 # height of the canvas
    
fig = plt.figure(figsize=(w,h))

myplot=fig.add_subplot(111)
myplot.plot_date(x=vx, y=vy, fmt="r-",color="#74c476", fillstyle="top")

myplot.set_title(graph_title, fontstyle='italic', fontsize=18)
myplot.set_ylabel("Number of Tweets")
myplot.grid(True,linestyle='-',color='0.75')

fig.autofmt_xdate() # Auto-adjust and beautify the labels
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import os, csv, json
from time import time 
import datetime
from collections import Counter
import lib.tweetminer as minetweet
from lib.users import UserAPI
import networkx as nx
import community
from lib.nlp import NLPMiner
import locale

results_path="/home/clemsos/Dev/mitras/results/"
meme_names=["biaoge"]
# meme_names=[ meme for meme in os.listdir(results_path) if meme[-3:] != "csv"]
# meme_names=[
#  'biaoge',
#  'thevoice',
#  'moyan',
#  'hougong',
#  'gangnam',
#  'sextape',
#  'dufu',
#  'ccp',
#  'yuanfang',
#  'qiegao']


print meme_names


t0=time()
minetweet.init_tweet_regex()


locale.setlocale(locale.LC_ALL, "")

nlp=NLPMiner()

stoplist=[i.strip() for i in open("lib/stopwords/zh-stopwords","r")]
stoplist+=[i.strip() for i in open("lib/stopwords/stopwords.txt","r")]
stoplist+=["转发","微博","说 ","一个","【 ","年 ","转 ","请","＂ ","问题","知道","中 ","已经","现在","说","【",'＂',"年","中","今天","应该","真的","月","希望","想","日","这是","太","转","支持"]
# stoplist+=["事儿","中国"]


api=UserAPI()
words_users_time=[]

def get_province(_userid):
    province_code= api.get_province(_userid)
    # print province_code
    try :
        return api.provinces[province_code]
    except KeyError :
        return 0
        pass
```

```python
meme_name="biaoge"

# Init
tstart=time()
print "Processing meme '%s'"%meme_name

# files names
meme_path=outfile=results_path+meme_name
meme_csv=meme_path+"/"+meme_name+".csv"

jsondata={}
jsondata["meme_name"]=meme_name

user_edges=[]
user_nodes=[]
words_series=[]
words_list=[]
words_users={}
count=0

words_time={}
user_edges_time=[]
words_users_time={}

# process the data
with open(meme_csv, 'rb') as csvfile:
    memecsv=csv.reader(csvfile)
    memecsv.next() # skip headers

    for row in memecsv:
        # extract text
        t=row[1]    
        count+=1

        
        d=datetime.datetime.strptime(row[9], "%Y-%m-%dT%H:%M:%S")
        day = datetime.datetime(d.year,d.month,d.day,d.hour,0,0)
        timestamp=day.strftime("%s")

        # regexp extract tweet entities
        mentions,urls,hashtags,clean=minetweet.extract_tweet_entities(t)
        
        # extract text 
        dico=nlp.extract_dictionary(clean)

        # remove stopwords and get clean dico
        clean_dico=nlp.remove_stopwords(dico)
        
        # remove more stopwords
        tmp_words=[w for w in clean_dico if w.encode('utf-8') not in stoplist and w[0] != "u" ]

        words_list+=tmp_words # global list for counter            

        # User diffusion graph
        user_diff=[]
        for mention in mentions:
            user_edges.append((row[0],mention))
            user_edges_time.append((row[0],mention,timestamp))
            if mention not in user_diff : user_diff.append(mention)

            # retweeted_uid
        if row[7] != "" : 
            user_edges.append((row[7],row[0]))
            user_edges_time.append((row[7],row[0],timestamp))
            if row[7] not in user_diff : user_diff.append(row[7])


        words_series.append(tmp_words)

        for w in tmp_words :
            try :
                words_users[w] += user_diff
                words_time[w].append(timestamp)
                words_users_time[w].append((user_diff,timestamp))
            except KeyError:
                words_users[w]=[]
                words_users[w]+=user_diff
                
                words_time[w]=[]
                words_time[w].append(timestamp)
                
                words_users_time[w]=[]
                words_users_time[w].append((user_diff,timestamp))
```

```python
word=node_words[1]

counter={}
for exchange in words_users_time[word["name"]]:
    if exchange[0]!=[]: 
        try :
            counter[exchange[1]]+=exchange[0]
        except KeyError:
            counter[exchange[1]]=[]
            counter[exchange[1]]+=exchange[0]

a=[(t,Counter(counter[t]).most_common()) for t in counter]

words_province_by_time=[]
users_by_time=[]
for timeframe in a:
    
    
    # users
    u={}
    u["time"]=timeframe[0]
    u["users"]=[{"user": c[0], "count":c[1]} for c in timeframe[1]]
    users_by_time.append(u)
    
    # provinces
    t={}
    t["time"]=timeframe[0]
    
    for usercount in timeframe[1]:
        prov=[]
        user=usercount[0]
        
        province=get_province(user)
        prov.append(province)
        t["provinces"]=[{"province": c[0], "count":c[1]} for c in Counter(prov).most_common()]
        
    words_province_by_time.append(t)

word["provinces_by_time"]=words_province_by_time
word["users_by_time"]=users_by_time

for key in word:
    print key
```
